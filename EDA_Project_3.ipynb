### Feature Understanding and ploting Charts

```python
plt.figure(figsize=(8,4))
sns.lineplot(data=df, x='Date', y='Total Amount', errorbar=None, color='purple')
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='mako', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['Quantity'], df['Total Amount'], color='orange', alpha=0.7, edgecolor='black')
plt.title('Quantity vs Total Amount')
plt.xlabel('Quantity')
plt.ylabel('Total Amount')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.hist(df['Total Amount'], bins=20, color='teal', edgecolor='black', alpha=0.7)
plt.title('Distribution of Total Sales Amount')
plt.xlabel('Total Amount')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
